## Improving SGD for logistic regression

```r
# import functions
source('~/Box Sync/PhDCourses/SDS385Statistical_models_for_big_data/SDS385/solutions/exercises04/improved_SGD.R')
```

```r
# generate simulated data
set.seed(444)
x1 = rnorm(1000)  # some continuous variables 
x2 = rnorm(1000)
x = cbind(as.matrix(x1), as.matrix(x2))
x= cbind(x, rep(1,1000))  # add one column for intercept
z = 1 + 2*x[ ,1] + 3*x[ ,2]  # linear combination with a bias
prob = 1/(1+exp(-z))  # pass through an inv-logit function
y = as.matrix(rbinom(1000,1,prob))
```

### Part A

```r
# experiments for minibatch method
betas = matrix(1, 3, 1)
H = diag(dim(betas)[1])
ls.results <- LineSearchSGD(x, y, betas, H, max.iter = 30000, replace = T, refresh.rate = 100, batch.size = 2000, lambda=0.1)
```

```r
loglikelihood = ls.results[[3]]
plot(1:30000, loglikelihood, type='l', xlab='iteration', ylab='avg log likelihood')
```

### Part B

```r
# experiments for adaptive gradient method
betas = matrix(1, 3, 1)
ada.results <- AdaGradSGD(x, y, betas, step.size = 0.5, max.iter = 30000, replace = T, lambda=0.1)
```

```r
loglikelihood = ada.results[[3]]
plot(1:30000, loglikelihood, type='l', xlab='iteration', ylab='avg log likelihood')
```
